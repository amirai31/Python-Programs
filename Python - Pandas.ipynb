#### Q.1 Write a Pandas program to add, subtract, multiple and divide two Pandas Series.

```python
import pandas as pd
```

```python
series1 = pd.Series([2, 4, 6, 8, 10]) 
series2 = pd.Series([1, 3, 5, 7, 9])

print("Add two Series:")
print(series1 + series2)
print()

print("Subtract two Series:")
print(series1 - series2)
print()

print("Multiply two Series:")
print(series1 * series2)
print()

print("Divide Series1 by Series2:")
print(series1 / series2)
```

#### Create a Pandas Data frame from the given data and create a new column “Voter” based on voter age, i.e., if age>18 then voter column should be “Yes” otherwise if age <18 then voter column should be “No”

```python
#data
raw_Data = {'Voter_name': ['Geek1', 'Geek2', 'Geek3', 'Geek4',
'Geek5', 'Geek6', 'Geek7', 'Geek8'],
'Voter_age': [15, 23, 25, 9, 67, 54, 42, np.NaN]}

#creating new dataframe
df = pd.DataFrame(raw_Data)

#creating new column "Voter"
df['Voter'] = np.where(df['Voter_age']>18, "Yes", "No")

#display
df
```

#### Create a Pandas Data frame from the given data and collapse First and Last column into one column as Full Name, so the output contains Full Name and Age, then convert column age to index

```python
#data
raw_Data = {'First': ['Manan ', 'Raghav ', 'Sunny '],
'Last': ['Goel', 'Sharma', 'Chawla'],
'Age' : [12, 24, 56]}

#creating new dataframe
df = pd.DataFrame(raw_Data)

#creating new column "Full Name"
df['Full Name'] = df['First']+ ' ' + df['Last']

#display 
df
```

#### Create a Pandas Data frame from the given data 

```python
#data
raw_Data = {'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/2011'],
'Product':['Umbrella', 'Matress', 'Badminton','Shuttle'],
'Price':[1250, 1450, 1550, 400],
'Expense': [ 21525220.653, 31125840.875, 23135428.768, 56245263.942]}

#creating new dataframe
df = pd.DataFrame(raw_Data)

#display 
df
```

#### a- Add Index as Item1, Item2, Item3, Item4

```python
#adding the new index
df.index = ['Item1', 'Item2', 'Item3', 'Item4']

#display 
df
```

#### b- Find the index labels of all items whose ‘Price’ is greater than 1000.

```python
#finding the index labels
index_labels = df[df['Price']>1000].index.tolist()

#display 
index_labels
```

#### c- Replace products using map() with respective codes- Umbrella : ‘U’, Matress : 'M', Badminton: 'B', Shuttle: 'S'

```python
#new codes
codes =  {'Umbrella' : 'U', 'Matress' : 'M', 'Badminton': 'B', 'Shuttle': 'S'}

#using map() to replace the product names
df['Product'] = df['Product'].map(codes)

#display 
df
```

#### d- Round off the Expense column values to two decimal places.

```python
#round off to 2 decimal places
df['Expense'] = df['Expense'].round(2)

#display 
df
```

#### e- Create a new column called ‘Discounted_Price’ after applying a 10% discount on the existing ‘price’ column.(try using lambda function)

```python
#creating new column
df['Discounted_Price'] = df['Price'].apply(lambda x:(x*90)/100)

#display 
df
```

#### f- Convert the column type of “Date” to datetime format

```python
#datatype of Date column
df['Date'].dtype

#convert the column type
df['Date'] = pd.to_datetime(df['Date'])

#datatype of Date column
df['Date'].dtype
```

#### g- Create a column rank which ranks the products based on the price (one with the highest price will be rank 1).

```python
#creating new column
df['Rank'] = df['Price'].rank(ascending = False).astype(int)

#display 
df
```
